```python
# Вибір ключових фіч
selected_features = {
    'ЗАРПЛАТА / СУМАРНИЙ ДОХІД в IT у $$$ за місяць, лише ставка \nЧИСТИМИ - після сплати податків': 'salary_usd',
    'Тайтл': 'title',
    'Категорія': 'category',
    'Почніть вводити і оберіть вашу ОСНОВНУ посаду зі списку': 'position',
    'Вкажіть вашу основну спеціалізацію': 'specialization',
    'Основна мова програмування': 'main_language',
    'Знання англійської мови': 'english_level',
    'Загальний стаж роботи за нинішньою ІТ-спеціальністю': 'it_experience_years',
}

# Створити копію датафрейму лише з цими фічами та перейменувати колонки
df_prepared = df[list(selected_features.keys())].copy()
df_prepared.rename(columns=selected_features, inplace=True)
df_prepared.fillna({'main_language':'Unknown'}, inplace=True)
df_prepared.info()
```

```python
df_prepared = df_prepared.dropna(subset=['salary_usd'])
df_prepared.info()
```

```python
df_prepared['title'].value_counts()
```

```python
# трансформація і покарщення деяких категорій
category_map = {
    'SE': 'Software Engineer',
    'QA': 'Quality Assurance',
    'DS': 'Data Scientist',
    'Product Mgmt': 'Product Management',
    'Project Mgmt': 'Project Management',
    'Other Mgmt': 'Other Management'
}

df_prepared['category'] = df_prepared['category'].replace(category_map)
df_prepared['category'].value_counts()
```

```python
import re

# групування ідентичний або рідкісних позицій (для зменшення шуму)

def normalize_position(pos):
    pos = pos.lower()
    # Групування за ключовими словами
    if 'developer' in pos or 'розробник' in pos:
        return 'Developer'
    if 'qa' in pos or 'quality assurance' in pos or 'тестувальник' in pos:
        return 'QA'
    if 'analyst' in pos or 'аналітик' in pos:
        return 'Analyst'
    if 'manager' in pos or 'менеджер' in pos:
        return 'Manager'
    if 'devops' in pos:
        return 'DevOps'
    if 'designer' in pos or 'дизайнер' in pos:
        return 'Designer'
    if 'architect' in pos or 'архітектор' in pos:
        return 'Architect'
    if 'data scientist' in pos or 'data science' in pos:
        return 'Data Scientist'
    if 'support' in pos or 'підтримка' in pos:
        return 'Support'
    if 'admin' in pos or 'адміністратор' in pos:
        return 'Admin'
    return pos.strip().title()

# нормалізація позицій
df_prepared['position_grouped'] = df_prepared['position'].astype(str).apply(normalize_position)

# Рідкісні позиції (менше 50) замінимо на "Other"
counts = df_prepared['position_grouped'].value_counts()
rare_positions = counts[counts < 50].index
df_prepared['position_grouped'] = df_prepared['position_grouped'].replace(rare_positions, 'Other')
```
